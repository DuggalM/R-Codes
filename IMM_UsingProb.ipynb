```python
# Function for setting market segments
def market_segment(df):
    '''
    This function takes a dataframe and assigns the market segment to it that is used in the GGHM.
    
    Arguments: dataframe
    
    return: dataframe with market segment
    '''
    
    if {'hhinc', 'auto_suff'}.issubset(df.columns):
        
        # create segments
        df.loc[(df['hhinc'] <= 60000) & (df['auto_suff'] == 0), 'market_seg'] = 0
        df.loc[(df['hhinc'] > 60000) & (df['auto_suff'] == 0), 'market_seg'] = 1
        df.loc[(df['hhinc'] <= 60000) & (df['auto_suff'] == 1), 'market_seg'] = 2
        df.loc[(df['hhinc'] > 60000) & (df['auto_suff'] == 1), 'market_seg'] = 3
        df.loc[(df['hhinc'] <= 60000) & (df['auto_suff'] == 2), 'market_seg'] = 4
        df.loc[(df['hhinc'] > 60000) & (df['auto_suff'] == 2), 'market_seg'] = 5
        # set dtype
        df['market_seg'] = df['market_seg'].astype('int8')
    else:
        print("The requisite fields are not there to run the function")
    
    return(df)
```

### Functions to assign vehicle trips

```python
def veh_type_assign(df, segment, sampledf, sampledf_all_mseg, veh_type):
    
    '''
    This function takes the first trip purpose (hbw generally) and assigns a veh type to each household
    
    Args: The df that needs to be assigned a vehicle type
          Segment counter
          Dictionary name to collect DFs created for each zone within a market segment
          Dictionary name to collect DFs across zones and market segments
          File that has probabilities for the four veh type and 6 market segments for every zone
    
    Return: a dataframe that is the same length as the trip purpose df with veh type assigned to it
    
    '''

    # subset the groups by market segment and sort. Then create
    # the column indexes to slice the vehicle type dataframe.
    # The melt the dataframe to get it ready for sampling 
    g1 = df.loc[df['market_seg'] == segment].sort_values('taz')
    g2 = g1.groupby(['taz', 'market_seg']).size().reset_index(name = 'counts')

    # get column indices to slice
    segment = int(float(segment))
    s_pos = segment*4
    e_pos = segment*4+4
    col_ind = list(range(s_pos, e_pos))
    veh_type1 = veh_type.iloc[:, col_ind].reset_index()

    # melt the dataframe to be sampled from.
    veh_melt1 = pd.melt(veh_type1, id_vars = 'ggh_zone').sort_values('ggh_zone')

    # get unique zones to loop over
    uniq_zones = g1['taz'].unique()

    for zone in uniq_zones :

        # sample to produce the required records from the distribution in the zone of question
        # get the weights to sample from
        wts = veh_melt1.loc[veh_melt1['ggh_zone'] == zone]
        wts = wts['value']

        # number of observations to be sampled
        no_sam = g2.loc[g2['taz'] == zone]
        no_sam = int(float(no_sam.iloc[0]['counts']))

        # sample with replacement and collect each individual zones results in a dictionary
        sdf = veh_melt1.sample(n = no_sam, weights = wts, replace = True, random_state = seed)
        columns = ['ggh_zone', 'value']
        sdf.drop(columns, axis = 1, inplace = True)

        # concatenate to the appropriate households within the zone and belonging to the market segment
        # in question
        sdf.reset_index(drop=True, inplace=True)
        hh_df_zone = df.loc[(df['taz'] == zone) & (df['market_seg'] == segment) ]
        hh_df_zone.reset_index(drop=True, inplace=True)
        sdf = pd.concat([hh_df_zone, sdf], axis = 1)
        
        # collect in dictionary
        sampledf[zone] = sdf

    # now concatenate across segments
    sampledf_all_mseg[segment] = pd.concat(sampledf.values(), ignore_index = True)
    
        # run check            
    try:
        assert len(sampledf_all_mseg[segment]) == len(g1)

    except AssertionError as e:
        e.args += ('The length of the sampled dataframe is wrong.')
        raise

    # get master dataframe and cbind it to the hbw_trips df
    veh_type_df = pd.concat(sampledf_all_mseg.values(), ignore_index = True)
    veh_type_df.rename(columns = {'variable': 'veh_type'}, inplace = True)

    return(veh_type_df)
```

```python
# The trips_out file contains a peak hour factor column that decides whether a trip is sampled
# in the peak or off-peak period. In order to discretely select the peak records and vice-versa
# an uniform random number generator is run and the values are attached to the trips_out file.
# If the (1-peak_factor) value in the record is greater than that of the random value than 
# the record is in the off-peak and vice-versa.

np.random.seed(seed)
random = pd.DataFrame(np.random.uniform(size = len(trips)))
random.columns = ['rnum']

# attach the random number generator and calculate peak_flag. A value of 1 in this flag
# means that this is a peak period trip record.
trips = pd.concat([trips, random], axis = 1)
trips['peak_flag'] = np.where((1 - trips['peak_factor']) > trips['rnum'], 0, 1)
trips[['peak_flag']] = trips[['peak_flag']].astype('int8')   # save some memory
```
